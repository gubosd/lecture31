### 데이터 불러오기

- 대부분의 데이터는 표형태의 텍스트파일이다
- iris.csv 파일을 불러온다


```python
import pandas as pd

data = pd.read_csv('iris.csv')
data
```

```python
iris = data.iloc[:, :4].values # to_numpy()
iris
```

```python
display(type(iris), iris.dtype, iris.shape, iris.ndim)
```

```python
iris.tolist()
```

### 정수/실수 타입 변환
- 머신러닝에서는 실수를 기본타입으로 사용하므로 정수 -> 실수로 변환이 필요하다

```python
a = np.arange(12).reshape(4,3)
a
```

```python
b = np.array(a, dtype='float')
b
```

```python
a.dtype, b.dtype
```

```python
b = a.astype('float')
b
```

```python
a = np.arange(1, 3, 0.1)
a
```

```python
b = a.astype('int')
b
```

### 값 바꾸기

```python
a = np.arange(25).reshape(5,5)
a
```

```python
a[0,0] = -1
a
```

```python
a[:, -1] = -2
a
```

```python
a[4] = -3
a
```

```python
b = a[1:4, 1:4] # b 는 새로 생성되지 않고 a 를 참조하고 있다
b[1,2] = 100    # a 의 값이 바뀐다
a
```
