### 통계를 이용한 데이터 추출

```python
# 평균이 3 보다 큰 샘플만 뽑는다
iris[iris.mean(axis=1)>5]
```

```python
# 표준편차가 2.3 보다 큰 샘플만 뽑는다
iris[iris.std(axis=1)>2.3]
```

```python
# 평균이 4.5보다 크고, 표준편차가 2보다 작은 샘플들
iris[(iris.mean(axis=1)>4.5) & (iris.std(axis=1)<2)]
```

### 브로드캐스팅
- 두 개의 데이터가 모양(shape)이 다를때, 연산이 가능한가?
- 일반규칙 : https://numpy.org/doc/stable/user/basics.broadcasting.html

```python
a = np.zeros([4,3])
a
```

```python
a + [1,2,3]
```

```python
a + [[1],[2],[3],[4]]
```

```python
iris - iris.mean(axis=0) # 각 속성의 평균값을 구한 후, 이를 원본데이터에서 뺀 결과
```

```python
iris - iris.mean(axis=1).reshape(150,1) # 각 샘플의 평균값을 구한 후, 이를 원본데이터에서 뺀 결과
```

```python
iris_norm = (iris - iris.mean(axis=0)) / iris.std(axis=0) # 정규화 (평균 0, 표준편차 1로 변환)
iris_norm
```

```python
iris_norm.shape
```

```python
iris_norm.mean(axis=0), iris_norm.std(axis=0)
```

### 정렬
- np.sort()
- np.argsort()

```python
a = np.random.randint(1,7, size=10)
a
```

```python
np.sort(a)
```

```python
np.sort(a)[::-1]
```

```python
ids = np.argsort(a) # 각 샘플의 인덱스를 출력
ids
```

```python
a[ids] # np.sort(a) 와 같은 결과임
```

```python
np.argmax(a)
```

```python
np.argmin(a)
```

- 2차원 표형태의 데이터를 정렬할 때는 np.argsort() 를 사용한다

```python
a = np.random.randint(1,7, size=[10,5])
a
```

```python
a[np.argsort(a[:,0])] # 첫번째 열을 기준으로 정렬
```

```python
a[np.argsort(a[:,-1])] # 마지막 열을 기준으로 정렬
```

```python
iris[np.argsort(iris.mean(axis=1))] # 각 샘플의 평균값이 작은 순서대로 정렬
```

```python
iris[np.argsort(iris.std(axis=1))] # 각 샘플의 평균값이 작은 순서대로 정렬
```

### 데이터 합치기
- np.r_[a,b] # 행으로 늘리기
- np.c_[a,b] # 열쪽으로 늘리기

```python
a = np.random.randn(10,3)
b = np.random.randn(5,3)

np.r_[a,b]
```

```python
a = np.random.randn(10,3)
b = np.random.randn(10,2)

np.c_[a,b]
```

```python
iris2 = np.c_[iris, iris.mean(axis=1), iris.std(axis=1)]
iris2.shape
```

```python
iris3 = np.r_[iris, iris.mean(axis=0).reshape(1,4)]
iris3.shape
```

### 수학함수 그리기
